```python
from os import path
from pydub import AudioSegment
from tqdm import tqdm
from os import listdir
```

```python
# Transformation
def change_format(mp3_file) :
    try :
        src = mp3_file
        pokemon = mp3_file.split("MP3\\")[1].split(".mp3")[0]
        dst = r"C:\Users\ibiza\OneDrive\Desktop\Cours\Python\Time Series\Son\Data\WAV"+'\\'+pokemon+".wav"

        # convert wav to mp3                                                            
        sound = AudioSegment.from_mp3(src)
        sound.export(dst, format="wav")
    except:
        print(pokemon)
```

```python
#Liste des mp3s
liste_mp3s = listdir(r"C:\Users\ibiza\OneDrive\Desktop\Cours\Python\Time Series\Son\Data\MP3")
liste_mp3s = [r"C:\Users\ibiza\OneDrive\Desktop\Cours\Python\Time Series\Son\Data\MP3"+'\\'+name 
             for name in liste_mp3s]
liste_mp3s
```

```python
for sound in tqdm(liste_mp3s) :
    change_format(sound)
```
