```python
#Arranger les pokemons par groupes de reproduction
def get_all_groupes() :
    egg_groups = {}
    for pokemon in groupes["Name"] :
        grp_pkm = groupes["Egg_Group_1"][groupes["Name"] == pokemon].item()
        if grp_pkm not in egg_groups :
            egg_groups[grp_pkm] = []
            egg_groups[grp_pkm].append(pokemon)
        else :
            egg_groups[grp_pkm].append(pokemon)
    return(egg_groups)
```

```python
egg_groups = get_all_groupes()
egg_groups
```

```python
#Arranger les pokemons par clusters
def get_all_classes() :
    pkm_classes = {}
    for pokemon in clusters["Pokemons"] :
        cluster_pkm = clusters["Cluster_kmeans"][clusters["Pokemons"] == pokemon].item()
        if cluster_pkm not in pkm_classes :
            pkm_classes[cluster_pkm] = []
            pkm_classes[cluster_pkm].append(pokemon)
        else :
            pkm_classes[cluster_pkm].append(pokemon)
    return(pkm_classes)
```

```python
pkm_classes = get_all_classes()
pkm_classes
```

```python
#Obtenir le grp / cluster d'un pokemon donné
def get_pokemon_grp_class(pokemon,Param,elements):
    if Param == "group" :
        dic = egg_groups
    elif Param == 'class' :
        dic = pkm_classes
    else :
        raise ValueError('Incorrect Parameter')
    for grp_class, liste_pokemon in dic.items():
        try :
            if pokemon in liste_pokemon:
                if elements == "label" :
                    return grp_class
                if elements == "content" :
                    return dic[grp_class]
        except Exception as e:
            print(e)
```

```python
#Pour un pokemon donné, Donne le nb de pokemons faisant 
#partis du meme egg group mais qui ne sont pas classé 
#dans le même cluster . Ils sont donc manquants par rapport
#A la classification officielle

def voisins_manquants(pokemon):
    nb_voisins_manquants = 0
    grp = get_pokemon_grp_class(pokemon,"group","label")
    classe = get_pokemon_grp_class(pokemon,"class","label")
    
    for other_pokemon in egg_groups[grp] :
        if other_pokemon not in pkm_classes[classe] :
            nb_voisins_manquants += 1 
    return nb_voisins_manquants
```

```python
#Pour un pokemon donné, Donne le nb de pokemons faisant 
#partis du meme cluster mais qui ne sont pas classé 
#dans le même egg group . Ils sont donc en trop par rapport
#A la classification officielle

def voisins_en_trop(pokemon):
    nb_voisins_en_trop = 0
    grp = get_pokemon_grp_class(pokemon,"group","label")
    classe = get_pokemon_grp_class(pokemon,"class","label")
    
    for other_pokemon in pkm_classes[classe]  :
        if other_pokemon not in egg_groups[grp] :
            nb_voisins_en_trop += 1 
    return nb_voisins_en_trop 
```

```python
def compute_all() :
    Pokedex_pourcentage = {}
    for pokemon in groupes["Name"] :
        #Nb d'erreurs
        x,y = voisins_manquants(pokemon),voisins_en_trop(pokemon)
        
        #On regarde les voisins de classe et grp uniques
        voisins_grp = get_pokemon_grp_class(pokemon,"group","content")
        voisins_classes = get_pokemon_grp_class(pokemon,"class","content")
        voisins_uniques = []
        voisins_uniques.extend(voisins_grp)
        voisins_uniques.extend(voisins_classes)
        voisins_uniques= list(set(voisins_uniques))
        
        nb_erreurs = (x+y) #Prendre en compte uniquement les voisins "importants"
        percentage_erreurs = nb_erreurs/(len(voisins_uniques)-1)
        
        Pokedex_pourcentage[pokemon] = [x,y,percentage_erreurs]                                        
    return Pokedex_pourcentage
```

```python
Pokedex_pourcentage = compute_all()
Pokedex_pourcentage
```

```python
def percentage_classe() :
    classe_percentage = {}
    for classe in pkm_classes.keys() :
        somme = 0
        liste_pokemons = pkm_classes[classe]
        for pokemon in liste_pokemons :
            somme = somme + Pokedex_pourcentage[pokemon][2]
        mean = somme/len(liste_pokemons)
        classe_percentage[classe] = mean
    return classe_percentage
```
