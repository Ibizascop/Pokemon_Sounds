```python
def get_all_groupes() :
    egg_groups = {}
    for pokemon in df["Name"] :
        grp_pkm = df["Groupe_1"][df["Name"] == pokemon].item()
        if grp_pkm not in egg_groups :
            egg_groups[grp_pkm] = []
            egg_groups[grp_pkm].append(pokemon)
        else :
            egg_groups[grp_pkm].append(pokemon)
    return(egg_groups)
```

```python
egg_groups = get_all_groupes()
egg_groups
```

```python
def get_all_classes() :
    pkm_classes = {}
    for pokemon in df["Name"] :
        class_pkm = #
        if class_pkm not in pkm_classes :
            pkm_classes[class_pkm] = []
            pkm_classes[class_pkm].append(pokemon)
        else :
            pkm_classes[class_pkm].append(pokemon)
    return(pkm_classes)
```

```python
pkm_classes = get_all_classes()
pkm_classes
```

```python
def get_pokemon_grp_class(pokemon,Param):
    if Param == "group" :
        dic = egg_groups
    elif Param == 'class' :
        dic = pkm_classes
    else :
        raise ValueError('Incorrect Parameter')
    for grp_class, liste_pokemon in dic.items():
        try :
             if pokemon in liste_pokemon:
                return grp_class
        except Exception as e:
            print(e)
```

```python
def voisins_manquants(pokemon):
    nb_voisins_manquants = 0
    grp = get_pokemon_grp_class(pokemon,"group")
    classe = get_pokemon_grp_class(pokemon,"class")
    
    for other_pokemon in egg_groups[grp] :
        if other_pokemon not in pkm_classes[classe] :
            nb_voisins_manquants += 1 
    return nb_voisins_manquants 
```

```python
def voisins_en_trop(pokemon):
    nb_voisins_en_trop = 0
    grp = get_pokemon_grp_class(pokemon,"group")
    classe = get_pokemon_grp_class(pokemon,"class")
    
    for other_pokemon in pkm_classes[classe]  :
        if other_pokemon not in egg_groups[grp] :
            nb_voisins_en_trop += 1 
    return nb_voisins_en_trop 
```
