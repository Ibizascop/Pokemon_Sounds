```python
#Import des différentes libreries
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
from IPython.display import Audio
```

```python
#Get all sounds links
#requete
req = requests.get("https://play.pokemonshowdown.com/audio/cries/")
page = BeautifulSoup(req.content,'html.parser')
print(page)

#traitement
table = page.find("table")
rows = table.findAll("tr")
sounds = [] 
for row in tqdm(rows[3:]) :
    try :
        sound = row.find("a")["href"]
        sounds.append(sound)
    except:
        pass
sounds = ["https://play.pokemonshowdown.com/audio/cries/"+sound for sound in sounds 
          if "mp3" in sound if "mega" not in sound ]
sounds
```

```python
#Obtenir et stocker les sons
for link in tqdm(sounds) :
    try : 
        pokemon = link.split("cries/")[1].split(".mp3")[0]
        req = requests.get(link)
        path = r'C:\Users\ibiza\OneDrive\Desktop\Cours\Python\Time Series\Son\Data\MP3'
        with open(path+'\\'+pokemon+'.mp3','wb') as file :
            file.write(req.content)
    except:
        print(pokemon)
```
